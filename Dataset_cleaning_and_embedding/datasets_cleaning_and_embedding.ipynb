```python
experience_map = {
    "fresher": "Fresher",
    "entry-level": "Fresher",

    "junior": "Junior",

    "mid-level": "Mid",
    "mid-level ": "Mid",
    "mid-level": "Mid",

    "mid-senior level": "Mid-Senior",
    "mid-senior": "Mid-Senior",

    "senior": "Senior",
    "senior-level": "Senior",
    "lead": "Senior",

    "experienced": "Experienced"
}

df_jobs["ExperienceLevel"] = df_jobs["ExperienceLevel"].map(experience_map)
```

```python
def get_experience_level(hours, jobs):

    score = hours + (jobs * 20)

    if score <= 200:
        return "Fresher"
    
    elif score <= 800:
        return "Junior"
    
    elif score <= 2500:
        return "Mid"
    
    elif score <= 6000:
        return "Senior"
    
    else:
        return "Experienced"


df_candiates["ExperienceLevel"] = df_candiates.apply(
    lambda row: row["ExperienceLevel"]
    if pd.notna(row["ExperienceLevel"])
    else get_experience_level(row["totalHours"], row["totalJobs"]),
    axis=1
)
```

```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    
    if not isinstance(text, str):
        return ""
    
    # lowercase
    text = text.lower()
    
    # remove unwanted characters but keep + # .
    text = re.sub(r"[^a-zA-Z0-9+#.\s]", " ", text)
    
    # remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()
    
    # tokenization
    tokens = word_tokenize(text)
    
    # remove stopwords
    tokens = [word for word in tokens if word not in stop_words]
    
    # lemmatization
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    # join back
    return " ".join(tokens)
```

```python
df_candiates['profile_text'] = df_candiates['profile_text'].apply(clean_text)
df_jobs['job_text'] = df_jobs['job_text'].apply(clean_text)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")

job_embeddings = model.encode(df_jobs["job_text"].tolist()).astype("float32")
candidate_embeddings = model.encode(df_candiates["profile_text"].tolist()).astype("float32")
```

```python
import faiss

faiss.normalize_L2(job_embeddings)
faiss.normalize_L2(candidate_embeddings)
```

```python
# for candidates embedding -> save to FAISS vector database
dimension = candidate_embeddings.shape[1]
candidate_index = faiss.IndexFlatIP(dimension)
candidate_index.add(candidate_embeddings)


faiss.write_index(candidate_index, "C:\\Users\\DELL\\Documents\\Skill sync\\embeddings\\candidate_index.faiss")
```

```python
# for job embeddings -> save to FAISS vector database
job_index = faiss.IndexFlatIP(dimension)
job_index.add(job_embeddings)

faiss.write_index(job_index, "C:\\Users\\DELL\\Documents\\Skill sync\\embeddings\\job_index.faiss")
```
